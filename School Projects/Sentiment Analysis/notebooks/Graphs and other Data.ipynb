```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.style.use('ggplot')


```

```python
amazon = pd.read_csv('amazon_pantry_analysis.csv')
amazon_cleaned = pd.read_csv('amazon_pantry_tweets_analysis_cleaned.csv')
```

```python
ax = sns.barplot(data=amazon, x='overall', y = 'neg')
ax.set_title('Amazon Pantry Negative VADER Score')
plt.show()
             
ax = sns.barplot(data=amazon, x='overall', y = 'pos')
ax.set_title('Amazon Pantry Positive VADER Score')
plt.show()
             
ax = sns.barplot(data=amazon, x='overall', y = 'neu')
ax.set_title('Amazon Pantry Neutral VADER Score')
plt.show()

ax = sns.barplot(data=amazon, x='overall', y = 'compound')
ax.set_title('Amazon Pantry Compound VADER Score')
plt.show()
             
ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'neg')
ax.set_title('Amazon Pantry Negative VADER Score-Cleaned')
plt.show()

ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'pos')
ax.set_title('Amazon Pantry Positive VADER Score-Cleaned')
plt.show()
             
ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'neu')
ax.set_title('Amazon Pantry Neutral VADER Score-Cleaned')
plt.show()
             
ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'compound')
ax.set_title('Amazon Pantry Compound VADER Score-Cleaned')
plt.show()
```

```python
ax = sns.barplot(data=amazon, x='overall', y = 'roberta_neg')
ax.set_title('Amazon Pantry Negative RoBERTa Score')
plt.show()

ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'roberta_neg')
ax.set_title('Amazon Pantry Negative RoBERTa Score-Cleaned')
plt.show()

ax = sns.barplot(data=amazon, x='overall', y = 'roberta_pos')
ax.set_title('Amazon Pantry Positive RoBERTa Score')
plt.show()

ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'roberta_pos')
ax.set_title('Amazon Pantry Positive RoBERTa Score-Cleaned')
plt.show()

ax = sns.barplot(data=amazon, x='overall', y = 'roberta_neu')
ax.set_title('Amazon Pantry Neutral RoBERTa Score')
plt.show()

ax = sns.barplot(data=amazon_cleaned, x='overall', y = 'roberta_neu')
ax.set_title('Amazon Pantry Neutral RoBERTa Score-Cleaned')
plt.show()







             
```
