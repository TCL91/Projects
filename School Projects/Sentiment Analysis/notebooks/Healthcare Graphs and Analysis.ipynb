```python
ax = sns.barplot(data=healthcare_group_ave, x='Number of Favorites', y = 'neg')
ax.set_title('Healthcare Negative VADER Average Score')
plt.show()
             
ax = sns.barplot(data=healthcare_group_ave, x='Number of Favorites', y = 'pos')
ax.set_title('Healthcare Positive VADER Average Score')
plt.show()
             
ax = sns.barplot(data=healthcare_group_ave, x='Number of Favorites', y = 'neu')
ax.set_title('Healthcare Neutral VADER Average Score')
plt.show()

ax = sns.barplot(data=healthcare_group_ave, x='Number of Favorites', y = 'compound')
ax.set_title('Healthcare Compound VADER Average Score')
plt.show()
             
ax = sns.barplot(data=healthcare_group_ave_cleaned, x='Number of Favorites', y = 'neg')
ax.set_title('Healthcare Negative VADER Average Score-Cleaned')
plt.show()

ax = sns.barplot(data=healthcare_group_ave_cleaned, x='Number of Favorites', y = 'pos')
ax.set_title('Healthcare Positive VADER Average Score-Cleaned')
plt.show()
             
ax = sns.barplot(data=healthcare_group_ave_cleaned, x='Number of Favorites', y = 'neu')
ax.set_title('Healthcare Neutral VADER Average Score-Cleaned')
plt.show()
             
ax = sns.barplot(data=healthcare_group_ave_cleaned, x='Number of Favorites', y = 'compound')
ax.set_title('Healthcare Compound Average VADER Score-Cleaned')
plt.show()
```
