```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.style.use('ggplot')


```

```python
education = pd.read_csv('healthcare_tweets_analysis.csv')
education_group_ave = pd.read_csv('education_group_ave.csv')

education_cleaned = pd.read_csv('education_tweets_analysis_cleaned.csv')
education_group_ave_cleaned = pd.read_csv('education_group_average_cleaned.csv')
```

```python
ax = sns.barplot(data=education_group_ave, x='Number of Favorites', y = 'neg')
ax.set_title('Education Negative VADER Average Score')
plt.show()
             
ax = sns.barplot(data=education_group_ave, x='Number of Favorites', y = 'pos')
ax.set_title('Education Positive VADER Average Score')
plt.show()
             
ax = sns.barplot(data=education_group_ave, x='Number of Favorites', y = 'neu')
ax.set_title('Education Neutral VADER Average Score')
plt.show()

ax = sns.barplot(data=education_group_ave, x='Number of Favorites', y = 'compound')
ax.set_title('Education Compound VADER Average Score')
plt.show()
             
ax = sns.barplot(data=education_group_ave_cleaned, x='Number of Favorites', y = 'neg')
ax.set_title('Education Negative VADER Average Score-Cleaned')
plt.show()

ax = sns.barplot(data=education_group_ave_cleaned, x='Number of Favorites', y = 'pos')
ax.set_title('Education Positive VADER Average Score-Cleaned')
plt.show()
             
ax = sns.barplot(data=education_group_ave_cleaned, x='Number of Favorites', y = 'neu')
ax.set_title('Education Neutral VADER Average Score-Cleaned')
plt.show()
             
ax = sns.barplot(data=education_group_ave_cleaned, x='Number of Favorites', y = 'compound')
ax.set_title('Education Compound Average VADER Score-Cleaned')
plt.show()
```
